# EDA

EDA (Exploratory Data Analysis - Keşifsel Veri Analizi), bir veri setini anlamak ve içindeki ilişkileri, desenleri ve anlamlı bilgileri ortaya çıkarmak için kullanılan bir süreçtir. EDA, veri setinin özelliklerini ve yapılarını keşfetmek, aykırı değerleri tanımlamak, eksik verileri ele almak, değişkenler arasındaki ilişkileri incelemek ve gelecek analizler ve modelleme için veri hazırlığı yapmak için kullanılır.

EDA'nın temel amacı, veri setinin içinde bulunan bilgiyi anlamak ve keşfetmektir. Bu, veri setinin genel özelliklerini anlamak için yapılan çeşitli görselleştirmeler ve istatistiksel analizler içerir. EDA sırasında, veri setinin dağılımını, merkezi eğilimini, değişkenler arasındaki ilişkileri ve olası örüntüleri görselleştirmek için histogramlar, kutu grafikleri, scatter plotlar gibi grafikler kullanılır.

## Korelasyon Matrisi

Korelasyon matrisini görselleştirir. Korelasyon matrisi, veri setindeki sayısal özellikler arasındaki ilişkileri gösteren bir matristir. Her bir hücre, iki özellik arasındaki korelasyon katsayısını temsil eder.

df.corr() yöntemi, DataFrame'deki sayısal sütunlar arasındaki korelasyon katsayılarını hesaplar ve bir korelasyon matrisi oluşturur.

sns.heatmap() fonksiyonu, korelasyon matrisini renkli bir ısı haritası olarak görselleştirir. Renkler, korelasyon katsayısının büyüklüğünü gösterir. cmap="coolwarm" parametresi, ısı haritasının renk paletini belirler. "coolwarm" paleti, hem pozitif hem de negatif korelasyonları görselleştirmek için kullanışlıdır.

Bu görselleştirme, veri setindeki sayısal özellikler arasındaki ilişkileri hızlıca anlamamıza yardımcı olur. Korelasyon matrisi, hangi özelliklerin birlikte değiştiğini ve birbirleriyle nasıl ilişkili olduğunu belirlememize yardımcı

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
import pylab
import scipy.stats as stats
```

MinMaxScaler: Veri setindeki sayısal özelliklerin ölçeklendirilmesi için kullanılır. Özellik değerlerini belirli bir aralığa ölçeklendirir, genellikle [0, 1] veya [-1, 1] aralığına.

OneHotEncoder: Kategorik değişkenleri sayısal değerlere dönüştürmek için kullanılır. Her bir kategori için bir sütun oluşturur ve bu sütunlara 0 veya 1 değerlerini atar.

LabelEncoder: Kategorik değişkenleri sayısal değerlere dönüştürmek için kullanılır. Her bir kategoriye bir tamsayı etiketi atar.

train_test_split: Veri setini rastgele iki alt kümeye böler. Önceden belirlenmiş bir oranda veri, eğitim setine atanırken geri kalanı test setine atanır. Bu, modelin eğitim sırasında görmediği veriler üzerinde performansını değerlendirmemize olanak tanır.

LinearRegression: Doğrusal regresyon, bağımsız değişkenlerle bağımlı değişken arasındaki ilişkiyi modellemek için kullanılır. LinearRegression sınıfı, veri setindeki bağımsız değişkenlerin (X) ve bağımlı değişkenin (Y) arasındaki ilişkiyi en iyi şekilde temsil eden bir doğrusal regresyon modelini oluşturur.

mean_squared_error: Gerçek ve tahmin edilen değerler arasındaki ortalama kare hatayı hesaplar. Bu, modelin tahminlerinin gerçek değerlerden ne kadar uzak olduğunu ölçer. Daha küçük bir ortalama kare hata, daha iyi bir model performansını gösterir.

mean_absolute_error: Gerçek ve tahmin edilen değerler arasındaki ortalama mutlak hatayı hesaplar. Bu, modelin tahminlerinin gerçek değerlere göre ne kadar hata yaptığını ölçer. Daha küçük bir ortalama mutlak hata, daha iyi bir model performansını gösterir.

math: matematiksel işlemler için kullanılır. Özellikle, math.sqrt() fonksiyonu karekök alma işlemi yapar.

pylab: matplotlib, numpy ve daha birçok kütüphaneyi içeren geniş bir modül setidir. Bu modül, çeşitli bilimsel ve mühendislik uygulamaları için kullanılır.

scipy.stats: istatistiksel hesaplamalar ve olasılık dağılımları gibi istatistiksel işlemler için bir araçlar koleksiyonu sağlar.

Bu modüller ve fonksiyonlar, veri setinin işlenmesi, modelin eğitilmesi ve performansının değerlendirilmesi için önemlidir.

```python
X = dataframe[["Avg. Session Length","Time on App", "Time on Website", "Length of Membership"]]
Y = dataframe["Yearly Amount Spent"]
```

X değişkeni, "Avg. Session Length" (Ortalama Oturum Süresi), "Time on App" (Uygulamada Geçirilen Zaman), "Time on Website" (Web Sitesinde Geçirilen Zaman) ve "Length of Membership" (Üyelik Süresi) özelliklerini içerir. Bu özellikler, modelin girdileri olarak kullanılacaktır.
Y değişkeni, "Yearly Amount Spent" (Yıllık Harcanan Miktar) özelliğini içerir. Bu özellik, modelin hedefi olarak kullanılacaktır, yani tahmin etmeye çalıştığımız değerdir.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
ol =  LinearRegression()
```

```python
ol.fit(X_train, Y_train)
```

```python
ol.coef_
```

 bir doğrusal regresyon modelinin katsayılarını (coefficients) temsil eden bir özelliktir. Bu katsayılar, bağımsız değişkenlerin (X) etkilerini belirtir ve modelin tahminini yaparken kullanılır.

```python
kts = pd.DataFrame(ol.coef_, X.columns, columns=["Katsayi"])
print(kts)
```

pd.DataFrame(ol.coef_, X.columns, columns=["Katsayi"]) ifadesi, ol.coef_ dizisini kullanarak bir DataFrame oluşturur. Bu işlemde, katsayılar (ol.coef_) X.columns ile belirtilen bağımsız değişken isimlerine (sütun adlarına) atanır. Son olarak, bu sütunlar "Katsayi" adı altında yeniden adlandırılır.

```python
tahminler = ol.predict(X_test)
```

```python
tahminler
```

Test setindeki bağımsız değişkenlerin (X_test) kullanılarak tahminlerin yapılmasını sağlar. Bu, modelin eğitim sürecinde öğrendiği ilişkileri kullanarak test setindeki girdiler için tahminler üretir.
Bu kodun çıktısı, test setindeki her bir girdi için bir tahmin değeri içeren bir dizi veya array olacaktır. Her bir tahmin, modelin test setindeki bağımsız değişkenler için ürettiği tahmin edilen bağımlı değişken değerini temsil eder.
bağımsız değişkenler için modelin yaptığı tahminleri yazdırdık. Bu tahminler, modelin test setindeki bağımsız değişkenler üzerindeki performansını değerlendirmek için kullanılabilir.


```python
sns.scatterplot(tahminler, Y_test)
plt.xlabel("Tahminler")
plt.ylabel("Gerçek Değerler")
plt.title("OL Modelinin Değerlendirmesi")
plt.show()
```

Bu, modelin performansını görsel olarak değerlendirmek için kullanılır.

sns.scatterplot(tahminler, Y_test) ifadesi, modelin tahminlerini (x-ekseni) ve gerçek değerlerini (y-ekseni) içeren bir saçılım grafiği oluşturur. Bu grafiğe her bir veri noktası temsil edilir. Tahmin edilen değer ile gerçek değer arasındaki ilişkiyi görmemizi sağlar.
plt.xlabel("Tahminler") ve plt.ylabel("Gerçek Değerler") ifadeleri, x-ekseni ve y-ekseninin etiketlerini belirler.
plt.title("OL Modelinin Değerlendirmesi") ifadesi, grafiğin başlığını belirler.
Bu grafiği kullanarak, modelin tahminlerinin gerçek değerlerle ne kadar uyumlu olduğunu görsel olarak değerlendirebiliriz. Eğer modelin tahminleri gerçek değerlerle iyi bir uyum içerisindeyse, grafiğin dağılımı genellikle çizgiye yakın bir şekilde olacaktır.

```python
print("Mean Absolute Error:", mean_absolute_error(Y_test, tahminler))
print("Mean Squared Error", mean_squared_error(Y_test, tahminler))
print("RMSE", math.sqrt(mean_squared_error(Y_test, tahminler)))
```

mean_absolute_error(Y_test, tahminler): Gerçek ve tahmin edilen değerler arasındaki ortalama mutlak hatayı hesaplar.

mean_squared_error(Y_test, tahminler): Gerçek ve tahmin edilen değerler arasındaki ortalama kare hatayı hesaplar.

math.sqrt(mean_squared_error(Y_test, tahminler)): Ortalama kare hatanın karekökünü alarak, kök ortalama kare hatayı (RMSE) hesaplar.

Bu değerler, modelin tahminlerinin ne kadar başarılı olduğunu gösterir. Daha küçük bir ortalama mutlak hata ve ortalama kare hata, daha iyi bir model performansını gösterir. Kök ortalama kare hata (RMSE), ortalama kare hatanın standart sapmasını ifade eder ve genellikle modelin tahminlerinin ne kadar doğru olduğunu daha iyi anlamamıza yardımcı olur.

```python
artik = Y_test - tahminler
```

```python
artik
```

Test setindeki gerçek değerlerle modelin tahminleri arasındaki farkı hesaplar. Bu işlem, her bir gözlem için bir artık değeri oluşturur. Eğer artık değeri negatif ise, model gerçek değerden daha düşük bir tahminde bulunmuş demektir. Eğer artık değeri pozitif ise, model gerçek değerden daha yüksek bir tahminde bulunmuş demektir.

```python
sns.displot(artik, bins = 20, kde=True)
```

Görselleştirmeyi kullanarak, modelin tahminlerinin gerçek değerlerden ne kadar uzak olduğunu daha iyi anlayabilirsiniz. Eğer artık değerleri normal bir dağılım gösteriyorsa, modelin tahminlerinin genel olarak dengeli olduğunu ve doğru tahminler yaptığını gösterebilir. Ancak, eğer artık değerleri farklı bir dağılım gösteriyorsa, modelin bazı durumlarda ne kadar hata yaptığını ve bu hataların nedenlerini daha detaylı olarak incelemek gerekebilir.

```python
stats.probplot(artik, dist = "norm", plot=pylab)
pylab.show()
```

stats.probplot(artik, dist="norm", plot=pylab):

pylab.show(): 

Oluşturulan olasılık plotunu gösterir.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Model
model = RandomForestRegressor()

# Hiperparametrelerin bir listesi
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Grid Search oluşturma
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Grid Search çalıştırma
grid_search.fit(X_train, Y_train)

# En iyi parametreler
print("En iyi parametreler:", grid_search.best_params_)

# En iyi skor
print("En iyi skor:", grid_search.best_score_)
```

"En iyi parametreler: {'max_depth': None, 'min_samples_leaf': 1, 'min_samples_split': 2, 'n_estimators': 150}": En iyi performansı sağlayan hiperparametre kombinasyonu belirlenir. "En iyi skor: 0.9371852210488434": Grid Search tarafından bulunan en iyi parametre kombinasyonu ile elde edilen ortalama çapraz doğrulama skoru yazdırılır. Bu skor, modelin belirli bir hiperparametre ayarında ne kadar iyi performans gösterdiğini gösterir.

RandomForestRegressor(n_estimators=1000, max_depth=30, min_samples_split=2, min_samples_leaf=1): Bu satırda, Random Forest Regressor modeli oluşturulur. n_estimators, max_depth, min_samples_split ve min_samples_leaf gibi hiperparametreler belirtilir. model.fit(X_train, Y_train): Model eğitilir. Eğitim verileri (X_train, Y_train) kullanılarak model, veri setindeki ilişkileri öğrenir.


model.predict(X_test): Eğitilmiş model, test verileri (X_test) üzerinde tahminler yapar. Tahminler, modelin veri setindeki örüntüleri kullanarak test verilerinin hedef değişkenlerini (Y_test) tahmin eder.


mean_squared_error(Y_test, predictions): Gerçek hedef değişkenlerle (Y_test) modelin tahminleri (predictions) arasındaki ortalama kare hatayı hesaplar. Daha küçük bir ortalama kare hata, modelin daha iyi performans gösterdiğini gösterir.


r2_score(Y_test, predictions): Gerçek hedef değişkenlerle (Y_test) modelin tahminleri (predictions) arasındaki varyans açıklama oranını hesaplar. Bu, modelin veri setindeki varyansın ne kadarını açıklayabildiğini gösterir. 1'e yakın bir R-kare değeri, modelin verilerdeki varyansın büyük bir kısmını açıkladığını gösterir.

"R-squared Score": 0.9481. Bu, modelin test setindeki varyansın %94.81'ini açıklayabildiğini gösteren R-kare skorudur. Yüksek bir R-kare değeri, modelin veri setindeki varyansın büyük bir kısmını açıkladığını gösterir.

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score


# Gradient Boosting Regressor modelini tanımlayalım
model = GradientBoostingRegressor()

# Denenecek hiperparametrelerin bir listesini oluşturalım
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# GridSearchCV ile hiperparametre ayarını gerçekleştirelim
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='r2', verbose=2, n_jobs=-1)
grid_search.fit(X_train, Y_train)

# En iyi parametreleri ve skoru yazdıralım
print("En iyi parametreler:", grid_search.best_params_)
print("En iyi skor:", grid_search.best_score_)

# En iyi modeli seçelim
best_model = grid_search.best_estimator_

# Test verisi üzerinde tahmin yapalım
predictions = best_model.predict(X_test)

# Test sonuçlarını değerlendirelim
mse = mean_squared_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)

print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

Gradient Boosting Regressor modelinin en iyi hiperparametre kombinasyonunu bulur ve bu kombinasyonla elde edilen en iyi modelin performansını değerlendirir. Sonuçlar, modelin test setindeki tahminlerinin gerçek değerlerle ne kadar iyi uyum sağladığını gösterir.
